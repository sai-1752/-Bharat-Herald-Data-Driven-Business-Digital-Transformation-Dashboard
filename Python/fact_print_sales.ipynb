```python

df_fact_print_sales['Copies Sold'] = pd.to_numeric(df_fact_print_sales['Copies Sold'], errors='coerce').fillna(0).astype(int)
df_fact_print_sales.info()
```

```python
import pandas as pd

df_fact_print_sales = pd.read_excel(
    "/Volumes/workspace/default/dataset/fact_print_sales.xlsx",
    dtype={"Month": str}
)

df_fact_print_sales["Copies Sold"] = (
    df_fact_print_sales["Copies Sold"]
    .astype(str)
    .str.replace(r"[^\d]", "", regex=True)
    .astype(int)
)

display(df_fact_print_sales)
```

```python
df_fact_print_sales['Language']=df_fact_print_sales['Language'].str.capitalize()
display(df_fact_print_sales)
```

```python
df_fact_print_sales['State']=df_fact_print_sales['State'].str.capitalize()
display(df_fact_print_sales)
```

```python
df_fact_print_sales['Month'] = pd.to_datetime(
    df_fact_print_sales['Month'],
    format='mixed',
    errors='coerce'
).dt.strftime('%Y-%m-%d')
display(df_fact_print_sales)
```

```python
df_fact_print_sales.rename(columns={'Month': 'Date', 'copies_returned': 'Copies_Returned','edition_ID': 'Edition_ID', 'City_ID': 'City_ID', 'Copies Sold': 'Copies_Sold', 'Net_Circulation': 'Net_Circulation'}, inplace=True)
```
